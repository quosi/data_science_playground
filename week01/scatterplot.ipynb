```python
#wandelt strings in integers
int_columns = [int(x) for x in fert.columns]
```

```python
#überschreibt Bezeichnung der Achsen mit den Jahreszahlen, die in int_columns gespeichert sind
fert.set_axis(axis=1, labels=int_columns, inplace=True)
```

```python
fert.columns
fert.head()
```

```python
#change into long format
#schiebt jahreszahlen-achse nach links
stack_fert = fert.stack()
stack_life = life.stack()
```

```python
#variables are type pd.Series now & can be converted to a DataFrame
data = {'fertility': stack_fert, 'lifeexp': stack_life}
df2 = pd.DataFrame(data)
df2.head(10)
```

```python
population = pd.read_excel('~/Projects/spiced/week1/plotting/gapminder_population.xlsx', index_col=0)
```
