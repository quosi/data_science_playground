```python
import requests
from bs4 import BeautifulSoup as soup
import re
```

```python
url = 'http://www.metrolyrics.com/best-coast-lyrics.html'
artist ='Best Coast'
#url = 'http://www.metrolyrics.com/rolling-stones-lyrics.html'
#url = 'http://www.metrolyrics.com/norah-jones-lyrics.html'
#url = 'http://www.metrolyrics.com/dream-theater-lyrics.html'
```

```python
def lyrics_scraper(url, artist):
    page = requests.get(url)
    html = soup(page.text, 'html.parser')
    result = html.find_all(attrs={'class':'title'})
    
    links = []
    title = []
    file_names = []
    for each in result:
        if each.get('title'):
            links.append(each.get('href'))
            title.append(each.get('title'))
    # get rid of lyrics suggestions on the end of page
    title = title[:-15]
    links = links[:-15]    
    
    for each in title:
        print(each)
        short_string = re.findall(f'{artist} (.+) lyrics', each)[0]
        no_blank_str = re.sub("[^\d*\w*]", "_", short_string, re.DOTALL)
        clean_string = re.sub(("__+"), "_", no_blank_str, re.DOTALL)
        clean_string = clean_string.lower()
        file_names.append(clean_string)        
    
    for i in range(0,(len(links))):
        song_html = requests.get(links[i])
        song_html_bs = soup(song_html.text, 'html.parser')
        song_table = song_html_bs.find_all(attrs={'id':'lyrics-body-text'})
        file=open(f'{file_names[i]}.txt',"w")
        print(f'scraping {file_names[i]}')
        for each in song_table[0].find_all('p'):
            file.write(each.text)
        #    print(each.text)
        file.close()
```

```python
lyrics_scraper(url, artist)
```
