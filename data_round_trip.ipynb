# Data Round-Trip

```python
import pandas as pd
```

## CSV

Read the CSV file `large_countries_2015.csv`

```python
countries=pd.read_csv("large_countries_2015.csv")
countries
```

## Excel

* Write the data into an Excel file
* Read the data again

```python
countries.to_excel("countries.xlsx", sheet_name='Sheet1')
countries2=pd.read_excel("countries.xlsx")
countries2
```

## JSON - Java Script Object Notation

* Write the data into a JSON file
* Read the data again

```python
countries.to_json("countries.json")
countries3=pd.read_json("countries.json")
countries3
```
