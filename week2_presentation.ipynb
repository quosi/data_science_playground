```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() #Set aesthetic parameters in one step.  

df=pd.read_csv("train.csv")
df=df.fillna(0)
df.head()
#-----------------------------------------------
d_age=pd.DataFrame(df[df['Survived']==0]['Age'])
l_age=pd.DataFrame(df[df['Survived']==1]['Age'])

# first way 
fig, (ax1,ax2,ax3) = plt.subplots(1,3, figsize=(10,4))
l_age.plot.hist(bins=15,color='blue',alpha=0.5,ax=ax1)
d_age.plot.hist(bins=15,color='red',alpha=0.3,ax=ax1)
ax1.legend(["Survive", "Dead"])

# second way 
df[df['Survived']==0]['Pclass'].plot.hist(bins=3,alpha=0.5,ax=ax2)
df[df['Survived']==1]['Pclass'].plot.hist(bins=3,alpha=0.5,ax=ax2)
ax2.legend(["Survive", "Dead"]) # adding legend 

# second way for gender (unstacked)
df.groupby('Survived')['Sex'].value_counts().unstack(1).plot.bar(stacked=True,alpha=0.5,ax=ax3)
# gender stacked
#df.groupby('Survived')['Sex'].value_counts().plot.bar(stacked=True,alpha=0.5,ax=ax3)
ax3.set_xticklabels(labels=['Female','Male'])
ax3.xaxis.label.set_visible(False) # 'xaxis.label.set.visible() '
ax3.legend().set_title('') #legend().set_title()
fig.tight_layout() # Fixing the spacing between figures
```

```python
# ix indexing works just the same as .loc when passed strings
df.ix[['Age']] == df.loc[['Age']]
# ix indexing works the same as .iloc when passed integers.
df.ix[[4]] == df.iloc[[4]]
# ix only works in both modes when the index of the DataFrame is NOT an integer itself.
```
