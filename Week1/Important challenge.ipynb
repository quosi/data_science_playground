```python
images = []
for y in range(1960, 2016):
    
    
    df6 = df3.unstack(1)
    df6 = df6[y]
    df6 = df6.unstack(1)
    # df6.plot.scatter('fertility', 'lifeexp', s=0.1)
    cmap = plt.get_cmap('tab20', lut = len(df6)).colors
    # df6.plot.scatter('fertility', 'lifeexp', s=0.1, c=cmap)
    df6.plot.scatter('fertility', 'lifeexp', s=df6['population'] // 1000000, c=cmap)
    print(f"Succesfully plotted for {y}")
    
    plt.title(f'Life Expectancy v. Fertility, Year: {y}')
    plt.axis((0,10,0,100))
    plt.savefig(f'frame_{y}.png')
    im = imageio.imread(f'frame_{y}.png')
    images.append(im)
    plt.figure()

imageio.mimsave('life_Expectancy.gif', images, fps=10)
```
